```python
df=pd.read_csv('C:\\Users\sunil\OneDrive\Documents\IMDb movies.csv', parse_dates=['date_published'], infer_datetime_format=True)
df.head()
```

```python
rdates=df[['title', 'year', 'country', 'date_published']]
rdates.rename(columns={'date_published':'date'}, inplace=True)
rdates
```

```python
r=rdates
r=r[r.title.str.contains('Christmas')]
r=r[r.country=='USA']
r.date.value_counts().sort_index().plot(kind='bar')
```

```python
r=rdates
r=r[r.title.str.contains('Christmas')]
r=r[r.country=='USA']
r.date.dt.month.value_counts().sort_index().plot(kind='bar')
```
